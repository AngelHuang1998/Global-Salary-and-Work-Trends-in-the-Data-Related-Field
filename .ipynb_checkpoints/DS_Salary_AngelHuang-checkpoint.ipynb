### Question 1: What are the percentages of each employment type?

````python
# Calculate the "proportion" of employment types
employment_ratio = df['employment_type'].value_counts(normalize=True).round(4)

# Convert labels into easy-to-read formats
label_dict = {
'FT': 'Full-time',
'PT': 'Part-time',
'CT': 'Contract employment',
'FL': 'Freelance'
}
labels = [label_dict.get(code, "Other") for code in employment_ratio.index] # If the corresponding index is not found, return "Other"

# Create a "conditional tag": only display tags with a percentage >= 1%, otherwise display an empty string to avoid crowding the graph
conditional_labels = [
    label if pct >= 0.01 else ''
    for label, pct in zip(labels, employment_ratio)
]

# Drawing
plt.figure(figsize=(4, 4))
plt.pie(
employment_ratio,
labels=conditional_labels,
autopct=lambda pct: f'{pct:.1f}%' if pct > 1 else '',  # The "percentage numbers" on the pie chart only show percentage values greater than 1%
colors= ['#B9A9C2', '#BFBC9D', '#9DA8AF', '#CBA29E'],
startangle=120,
)

plt.title('Employment Type', fontsize=14)
# Make sure the X and Y axes have the same scale so the pie chart looks round and not flattened.
plt.axis('equal')
plt.show()
````

**Result & Insight:**

## Feature Enrichment / Data Enrichment

### - Add a new column: `cost_index` (Cost of living index) by country
* Reason: To accurately assess real purchasing power, comparing salaries across countries alone can be misleading due to variations in living standards and the cost of goods. Therefore, the cost of living index is introduced as an adjustment factor to provide a more balanced comparison.
* Cost of living index source: https://www.numbeo.com/cost-of-living/

````python
# Cost of living index by country (base = 100 for the U.S.)
cost_index = {
    'US': 74.2,
    'CA': 68.0,
    'DE': 66.0,
    'GB': 63.6,
    'ES': 50.6,
    'IN': 24.4,
}

# Directly map cost_index into df based on company location
df['cost_index'] = df['company_location'].map(cost_index)

# Adjust the column order
cost_index_col = df.pop('cost_index')
df.insert(loc=4, column='cost_index', value=cost_index_col)

df.head(3)
````

### Question 7: For Data Engineers, which countries offer the top 5 highest average salaries? (Without consider the cost of living index) 
### Question 8: For Data Engineers, which countries offer the top 5 highest average "Real Purchasing Power"? (Consider the cost of living index) 
- `Real Purchasing Power = actual salary / (cost of living index / 100)`

````python
# Filter Data Engineers and calculate "Real Purchasing Power" adjusted by cost of living index
data_engineers = df[df['job_title_standardized'] == 'Data Engineer'].copy()
data_engineers['Real_Purchasing_Power'] = data_engineers['salary_in_usd'] / (data_engineers['cost_index'] / 100)


# Calculate the top 5 countries with the highest average salary (original and adjusted)
original_salaries = (
    data_engineers.groupby('company_location')['salary_in_usd']
    .mean().nlargest(5).sort_values()
)
adjusted_salaries = (
    data_engineers.groupby('company_location')['Real_Purchasing_Power']
    .mean().nlargest(5).sort_values()
)

# Plot side-by-side horizontal bar charts
fig, axes = plt.subplots(1, 2, figsize=(14, 5.5))

original_salaries.plot(kind='barh', ax=axes[0], color='#AAB8AB')
axes[0].set(title='Original Average Salary by Country for Data Engineers', 
            xlabel='Average Salary (Without consider the cost of living index)', ylabel='Country')

adjusted_salaries.plot(kind='barh', ax=axes[1], color='#C9B4C7')
axes[1].set(title='Average Real Purchasing Power by Country for Data Engineers', 
            xlabel='Average Real Purchasing Power (Consider the cost of living index)', ylabel='Country')

plt.tight_layout()
plt.show()
````

**Result & Insight:**
Countries with the highest original salaries do not necessarily retain their advantage after adjusting for cost of living, highlighting the significant impact of living expenses on real income.

## C. The relationship between remote work and salary

### Question 10: What is the remote work ratio distribution (0%, 50%, 100%) among the top 3 highest-paying jobs?

````python
# Identify the top 3 highest-paying job titles
top3_job = df.groupby('job_title_standardized')['salary_in_usd'].mean().sort_values(ascending=False).head(3).index

# Filter all records related to these three job titles
subset = df[df['job_title_standardized'].isin(top3_job)]

# Plot the distribution of remote_ratio for these 3 jobs(distinguished by color)
plt.figure(figsize=(6, 4.5))
sns.countplot(data=subset, x='job_title_standardized', hue='remote_ratio')
plt.title("Remote Work Ratio Distribution of Top 3 Highest-Paying Jobs")
plt.xlabel("Job Title")   
plt.ylabel("Count")       

plt.tight_layout()
plt.show()
````

**Result & Insight:**
Despite the remote-friendly nature of these high-paying roles, on-site work remains dominant—possibly due to better control over performance and security for tasks like model deployment and data processing within internal systems.

### Question 11: Among the top 3 most common jobs, what are the average salary differences between fully remote and fully on-site roles for entry-level (EN) employees? 

````python
# Find the top 3 most common job titles
top3_jobs = df['job_title_standardized'].value_counts().sort_values(ascending = False).head(3).index

# Filter out "Entry-Level (EN)" + "remote_ratio = 0 or 100" + "top3 jobs"
filtered_df = df[
    (df['experience_level'] == 'EN') &
    (df['job_title_standardized'].isin(top3_jobs)) &
    (df['remote_ratio'].isin([0, 100]))
]

# Calculate the average salary for each position under different remote_ratio
salary_by_remote_ratio = (
    filtered_df.groupby(['job_title_standardized', 'remote_ratio'])['salary_in_usd']
    .mean()
    .unstack()
    .round(2)
)

# Convert data to long format (for easy group bar chart drawing)
salary_melted = salary_by_remote_ratio.reset_index().melt(
    id_vars='job_title_standardized',
    value_vars=[0, 100],
    var_name='Remote Ratio',
    value_name='Average Salary (USD)'
)

# Drawing plot
plt.figure(figsize=(5.5, 4))
sns.barplot(
    data=salary_melted,
    x='job_title_standardized',
    y='Average Salary (USD)',
    hue='Remote Ratio',
    palette='muted'
)

plt.title("Entry-Level Salary Comparison: \nRemote vs On-Site in Top 3 Common Jobs")
plt.xlabel("Job Title")
plt.ylabel("Average Salary (USD)")
plt.legend(title="Remote Ratio (%)", fontsize=8.5, title_fontsize=9)

plt.tight_layout()
plt.show()
````

**Result & Insight:**
Among EN-level employees, fully on-site jobs generally offer higher average salaries than fully remote ones. This suggests that in-person work may help early-career professionals grow and command better pay.

## D. The relationship between company size and salary

### Question 12: How do job counts for Entry-Level (EN) roles differ between medium- and large-sized companies across job titles?

### Question 13: Among Entry-Level (EN) employees, how do average salaries vary between medium- and large-sized companies across different job titles?

* Based on the previous question, I exclude "Data Specialist" from this analysis due to insufficient data (only one record under company size M).

````python
# Filter data
filtered_en_df = df[
    (df['experience_level'] == 'EN') &
    (df['company_size'].isin(['M', 'L']))
]

# Group by job title and company size (M and L), calculate average salary
salary_by_title_size_en = (
     filtered_en_df
    .groupby(['job_title_standardized', 'company_size'])['salary_in_usd']
    .mean()
    .unstack()[['M', 'L']]   # Ensure column order: M → L
    .round(2)
)

# Reshape data for visualization (long format)
melted_salary = salary_by_title_size_en.reset_index().melt(
    id_vars='job_title_standardized',
    value_vars=['M', 'L'],
    var_name='Company Size',
    value_name='Average Salary (USD)'
)

# Plot horizontal bar chart
plt.figure(figsize=(8, 4.5))
sns.barplot(
    data=melted_salary,
    y='job_title_standardized',   
    x='Average Salary (USD)',       
    hue='Company Size',             
    palette='Set3'
)

plt.title("Average Salary for Entry-Level Roles:\nMedium vs Large Companies")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Job Title")
plt.legend(title="Company Size")

plt.tight_layout()
plt.show()
````

**Result & Insight:**
- "Data Specialist" shows outlier values, likely due to its small sample size, making it more susceptible to distortion. (only 4 records in large companies and 1 in medium companies )
- At the EN level, not all job titles show a trend of higher salaries in large companies, suggesting that the nature of the role and company-specific strategies may play a more significant role in determining compensation.

### Question 14: Among different company sizes, which job have the highest average salaries? (list Top 3 for each size)

````python
# Create an empty dictionary to store the top 3 highest-paying job for each company size
top3_jobs_by_company_size = {}

# Get the unique company size categories (S, M, L)
company_sizes = df['company_size'].unique()

# Process each company size separately
for size in company_sizes:
    subset = df[df['company_size'] == size]  # Filter records for the specific company size
    
    # Calculate the average salary for each job title in specific company size, and get the top 5
    job_salary = (
        subset.groupby('job_title_standardized')['salary_in_usd']
        .mean()
        .sort_values(ascending=False)
        .head(3)
    )
    
    # Store the result in the dictionary
    top3_jobs_by_company_size[size] = job_salary


# Display the results
for size, jobs in top3_jobs_by_company_size.items():
    print(f"\nCompany Size: {size}")
    print(jobs)
````

**Result & Insight:**
- Across all three company sizes, Data Researcher and ML Engineer consistently dominate the top-paying positions, indicating their high technical value and market scarcity.

- Medium-sized companies (M) tend to offer higher overall average salaries. The top 3 roles in this group all earn more than their counterparts in large and small companies. This may reflect a more aggressive compensation strategy to attract top talent.

- Small companies (S) generally offer more conservative salaries.

### Question 17: How did the proportion of fully remote work (remote_ratio = 100) change from 2020 to 2023?

````python
# Calculate the percentage of fully remote work (remote_ratio == 100) by year
full_remote_ratio_by_year = (
    df[df['remote_ratio'] == 100]
    .groupby('work_year')
    .size()
    .div(df.groupby('work_year').size())  # Divide by the total number of jobs in that year
    * 100
)

# line chart
plt.figure(figsize=(6, 3.2))
plt.plot(full_remote_ratio_by_year.index, full_remote_ratio_by_year.values, marker='o')
plt.title("Changes in Fully Remote Work Ratio (2020–2023)")
plt.xlabel("Year")
plt.ylabel("Fully Remote Ratio (%)")
plt.grid(True)
plt.xticks(full_remote_ratio_by_year.index) # set the X-axis ticks to specific position

plt.tight_layout()
plt.show()
````

**Result & Insight:**
* The year 2020 marked the peak of fully remote work, likely due to lockdown measures implemented during the COVID-19 pandemic.
* After the peak in 2020, the proportion of fully remote work declined year by year, suggesting that companies have been gradually shifting back to on-site or hybrid work models.

### Question 18: What is the percentage growth in job count for each job title from 2020 to 2023?

````python
# Count job occurrences by year and job title
job_year_counts = df.groupby(['work_year', 'job_title_standardized']).size().unstack()

# Extract job counts for 2020 and 2023
job_2020 = job_year_counts.loc[2020]
job_2023 = job_year_counts.loc[2023]

# Keep only jobs that existed in 2020 (to avoid divide-by-zero)
valid_jobs = job_2020[job_2020 > 0].index

# Compute growth rate
growth_rate = ((job_2023[valid_jobs] - job_2020[valid_jobs]) / job_2020[valid_jobs]) * 100

# Combine results into a new DataFrame
growth_df = pd.DataFrame({
    '2020 Count': job_2020[valid_jobs],
    '2023 Count': job_2023[valid_jobs],
    'Growth Rate (%)': growth_rate.round(2)
}).sort_values(by='Growth Rate (%)', ascending=True)


# horizontal bar chart
plt.figure(figsize=(7, 4.5))
sns.barplot(
    data=growth_df,
    x='Growth Rate (%)',
    y=growth_df.index,
    hue=growth_df.index,
    palette='viridis'
)
plt.title("Job Growth Rate from 2020 to 2023")
plt.xlabel("Growth Rate (%)")
plt.ylabel("Job Title")

plt.tight_layout()
plt.show()
````

**Result & Insight:**
Overall, data-related roles showed a growth trend from 2020 to 2023, indicating continued strength in the job market and strong future potential for the data field.

## F. Cross-Border Employment

### Question 19: Is cross-border employment common in the data field? 
 `cross-border employment : employee residence ≠ company location`

````python
cross_border_df = df[df['employee_residence'] != df['company_location']]

# Calculate the proportion of cross-border vs. local employment
cross_country_ratio = pd.Series({
    'Cross-Border': len(cross_border_df),
    'Local': len(df) - len(cross_border_df)
})

# Convert to percentage
cross_country_ratio = (cross_country_ratio / len(df) * 100).round(2)
cross_country_ratio
````

**Result & Insight:**
Cross-border employment is still not widespread, possibly due to factors such as tax regulations, time zone challenges, and data security concerns.

- (ES)  Spain
- (US)  United States

````python
# Calculate total hires and cross-border hires per country
total_hires = df['company_location'].value_counts()
cross_border_hires = cross_border_df['company_location'].value_counts()

# Combine into a DataFrame
cross_country_stats = pd.DataFrame({
    'total_hires': total_hires,
    'cross_border_hires': cross_border_hires
})

# Compute cross-border hire ratio (%)
cross_country_stats['cross_hire_ratio'] = (
    cross_country_stats['cross_border_hires'] / cross_country_stats['total_hires'] * 100
)

# Sort countries by cross-border ratio in descending order
cross_country_stats_sorted = cross_country_stats.sort_values(by='cross_hire_ratio', ascending=False)

# horizontal bar chart
plt.figure(figsize=(7, 4))
sns.barplot(
    data=cross_country_stats_sorted,
    x='cross_hire_ratio',
    y=cross_country_stats_sorted.index,
    hue=cross_country_stats_sorted.index, 
    palette='coolwarm'
)
plt.title("Cross-Border Employment Ratio by Country")
plt.xlabel("Cross-Border Employment Ratio (%)")
plt.ylabel("Company Location")

plt.tight_layout()
plt.show()
````

**Result & Insight:**
While cross-border employment remains uncommon overall, countries such as Germany (DE) and Canada (CA) stand out with relatively higher cross-border employment rates, indicating a more flexible approach to remote work in their data-related job markets.
